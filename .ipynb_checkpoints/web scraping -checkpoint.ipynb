## Here is code for web scraping

```python
import requests      #importing request modules
```

```python
url = " https://books.toscrape.com"
```

```python
response  = requests.get(url)
print(response)                # output =>200: OK (Success)

# print(response.text) it gives html scarped code
```

<h3> IT WILL SCARP WEB AND STORE HTML PAGES IN FOLDER "HTML pages of web"

```python
for i in range(1,51):
    a = requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    with open(f"HTML pages of web/bookpage-{i}.html","w",encoding="utf-8")as f:
        f.write(a.text)
    
    print(f"page {i} scraped succesfully")
```
