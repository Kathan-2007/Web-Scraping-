## DATA EXTRACTION USING BEAUTIFULSOUP


```python
import requests
from bs4 import BeautifulSoup
```

<h6>.
<h3>Opening and Reading html pages and finding all book name ,price ,rating and avalibility
    

```python
items = []
for i in range(1,51):
    with open(f"HTML pages of web/bookpage-{i}.html","r",encoding = "utf-8")as f:
        content = f.read()
        soup = BeautifulSoup(content,"html.parser")   #.html.parser is use to turns html data into easily search and extract data

       
      
        classarticle = soup.select("article.product_pod") 
        for article in classarticle:
            title = article.find("h3").find("a")["title"]                                 # it find titles of all books 
            
            price = article.select_one("p.price_color").text.split("£")[1]                # it find price of all books
            
            rating = article.select_one("p.star-rating")                                  # it find ratings of all books
            ratings = rating["class"][1]
            
            avalability  = article.select_one("p.instock.availability").get_text(strip=True)

            items.append([title,price,ratings,avalability])
```

## structuring data using pandas   

```python
import pandas as pd

df = pd.DataFrame(items,columns=["Book Name","Price","Star-rating","Avalibility"])
```

<h3> Storing data into "Books data.csv" file

```python
df.to_csv("Books data.csv",index=False)
```
