```python
import numpy as np

plt.hist(ages,bins=bins,edgecolor="black",linestyle='--',linewidth=1)
plt.axvline(np.mean(ages),color="skyblue",linestyle="-.",linewidth=2,alpha=1,label="Avrage Age")
plt.legend()
plt.title("Age Distribution with Mean Line")
```
