```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
 
plt.title("Career Runs of Indian Batsmen")

plt.pie(runs,labels=labels,colors=colors)
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor':'black','linewidth':1,'linestyle':'-.'}
)
```

```python
explode = [0.1,0,0,0]  # Only highlight Sachin's slice

plt.pie(runs, labels=labels, explode=explode, colors=colors,wedgeprops={'edgecolor':'black','linestyle':'-.','linewidth':1})
plt.title("Exploded Pie Example")
plt.show()

```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True  # adds a 3D-like shadow
)
```

```python
plt.pie(runs,labels=labels,explode=explode,colors=colors,shadow=True,startangle=120)
```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True,
    startangle=120,
    autopct='%1.1f%%')
```
